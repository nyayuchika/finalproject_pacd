#Segmentation

```python
#For segmentation using otsu thresholding
import cv2
img_otsu = []

for i in range(len(img_clahe)):
  # 1. Nilai ambang batas (threshold) yang dihitung oleh Otsu, kita beri nama 'thresh_value'
  # 2. Gambar hasil thresholding itu sendiri, kita beri nama 'segmented_image'
  # cv2.THRESH_BINARY: piksel > ambang batas menjadi 255 (putih), sisanya 0 (hitam)
  # cv2.THRESH_OTSU: flag untuk mengaktifkan metode Otsu
  thresh_value, segmented_image = cv2.threshold(img_clahe[i],0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
  img_otsu.append(segmented_image)
```

```python
plt.subplot(1,3,1)
plt.imshow(plt.imread(list_images[2]))
plt.title('Original Image')

plt.subplot(1,3,2)
plt.imshow(img_clahe[2], cmap = 'gray')
plt.title('CLAHE')

plt.subplot(1,3,3)
plt.imshow(img_otsu[2], cmap = 'gray')
plt.title('Otsu')
```

```python
img_flooded = []
# Suppose img is grayscale, lungs = dark, background = black
for i in range(len(img_otsu)):
  # 1. Pad image with 1-pixel white border
  padded = cv2.copyMakeBorder(img_otsu[i], 1, 1, 1, 1, cv2.BORDER_CONSTANT, value=255)
  # # 2. Flood-fill from top-left corner (now white border)
  mask = np.zeros((padded.shape[0]+2, padded.shape[1]+2), np.uint8)  # required size for floodFill
  cv2.floodFill(padded, mask, (0,0), 0)
  # # 3. Remove padding
  flooded = padded[1:-1, 1:-1]
  img_flooded.append(flooded)
```

```python
plt.subplot(1,3,1)
plt.imshow(plt.imread(list_images[2]))
plt.title('Original Image')

plt.subplot(1,3,2)
plt.imshow(img_otsu[2], cmap = 'gray')
plt.title('Otsu')

plt.subplot(1,3,3)
plt.imshow(img_flooded[2], cmap = 'gray')
plt.title('Flooded Fill')
```

```python
#Morpho close untuk menghilangkan noise di objek (nutup lubang paru-paru)
from scipy import ndimage
kernel_close = np.ones((11,11), np.uint8)
img_morpho_close = []

for i in range(len(img_flooded)):
  closing = cv2.morphologyEx(img_flooded[i], cv2.MORPH_CLOSE, kernel_close)
  lung_mask_filled = ndimage.binary_fill_holes(closing).astype(np.uint8) * 255
  num_labels, labels = cv2.connectedComponents(lung_mask_filled)
  sizes = [np.sum(labels == i) for i in range(1, num_labels)]
  largest = sorted(range(1, num_labels), key=lambda i: sizes[i-1], reverse=True)[:2]  # 2 lungs
  lung_mask_clean = np.isin(labels, largest).astype(np.uint8) * 255
  lung_mask_clean = cv2.medianBlur(lung_mask_clean, 5)  # smooth jagged edges
  img_morpho_close.append(lung_mask_clean)
img_morpho_close = np.asarray(img_morpho_close)
```

```python
# --- ASUMSI ---
# Anda sudah memiliki dua variabel ini dari tahap-tahap sebelumnya:

# 1. clahe_image: Gambar X-ray grayscale setelah di-CLAHE (dari preprocessing).
#    Ini adalah gambar sumber yang ingin kita "potong".
#
# 2. final_mask: Masker biner paru-paru yang sudah bersih (dari closing morphological).
#    Isinya hanya 0 (hitam) dan 255 (putih). Ini adalah "stensil" kita.
# ---

# Gunakan cv2.bitwise_and untuk mengisolasi area paru-paru
# Operasi ini akan mempertahankan piksel di 'clahe_image'
# hanya di lokasi di mana 'final_mask' berwarna putih (255).
# Di lokasi lain (di mana mask hitam), hasilnya akan menjadi 0 (hitam).

final_img = []
for i in range(len(img_morpho_close)):
    # --- Tahap 1: Membuat Masker Bersih (Kode Anda) ---
    # (Ini adalah kode Anda untuk membersihkan masker dari setiap gambar)
    image = img_morpho_close[i]
    lung_mask_filled = ndimage.binary_fill_holes(image).astype(np.uint8) * 255
    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(lung_mask_filled, connectivity=8)

    final_mask = np.zeros_like(image) # Mulai dengan masker hitam
    if num_labels > 1:
        areas = stats[1:, cv2.CC_STAT_AREA]
        sorted_indices = np.argsort(areas)[::-1]
        num_components_to_keep = min(2, len(sorted_indices))
        largest_indices = sorted_indices[:num_components_to_keep] + 1
        final_mask = np.isin(labels, largest_indices).astype(np.uint8) * 255
        final_mask = cv2.medianBlur(final_mask, 5)
    clahe_image = img_clahe[i]
    isolated_lungs = cv2.bitwise_and(clahe_image, clahe_image, mask=final_mask)
    final_img.append(isolated_lungs)

# --- HASIL ---
# Variabel 'segmented_lungs_grayscale' sekarang berisi gambar grayscale
# yang hanya menampilkan paru-paru dengan latar belakang hitam pekat.
#
# Gambar inilah yang siap untuk dinormalisasi ke rentang [0, 1]
# lalu diekstraksi fiturnya.
```

#Hitung IoU

```python
import numpy as np

def calculate_iou(mask_prediksi, mask_ground_truth):
    """
    Menghitung Intersection over Union (IoU) antara dua masker biner.

    Args:
    - mask_prediksi: NumPy array dari masker hasil segmentasi Anda (nilai 0 atau 255).
    - mask_ground_truth: NumPy array dari masker kunci jawaban (nilai 0 atau 255).

    Returns:
    - iou_score: Nilai IoU (float antara 0 dan 1).
    """
    # Pastikan kedua masker memiliki tipe data yang sama dan biner (0 atau 1)
    # untuk operasi logika. Kita ubah ke boolean (True/False).
    mask_pred_bool = mask_prediksi.astype(bool)
    mask_ground_truth_bool = mask_ground_truth.astype(bool)

    # Hitung Intersection (area di mana KEDUA masker True)
    intersection = np.logical_and(mask_pred_bool, mask_ground_truth_bool)

    # Hitung Union (area di mana SALAH SATU masker True)
    union = np.logical_or(mask_pred_bool, mask_ground_truth_bool)

    # Hitung jumlah piksel di intersection dan union
    intersection_area = np.sum(intersection)
    union_area = np.sum(union)

    # Hitung IoU, tambahkan epsilon (nilai kecil) untuk menghindari pembagian dengan nol
    iou_score = intersection_area / (union_area + 1e-6)

    return iou_score
```

```python
# ... (kode Anda yang lain) ...

# Asumsikan 'masks_list' adalah list berisi GAMBAR ground truth (NumPy array)
# yang sudah di-resize.
# Asumsikan 'img_morpho_close' adalah list berisi GAMBAR prediksi.

# Pastikan jumlahnya sama
if len(img_morpho_close) != len(masks_list):
    print("Error: Jumlah gambar prediksi tidak sama dengan jumlah gambar ground truth!")
else:
    iou_scores = []
    for i in range(len(masks_list)):
        # 1. Langsung ambil gambar ground truth dari list
        # HAPUS BARIS cv2.imread()
        ground_truth_mask = masks_list[i]

        # 2. Ambil masker prediksi ke-i
        predicted_mask = img_morpho_close[i]

        # 3. Panggil fungsi dengan DUA GAMBAR TUNGGAL
        score = calculate_iou(predicted_mask, ground_truth_mask)
        iou_scores.append(score)

        # print(f"IoU untuk gambar {i+1}: {score:.4f}")

    # ... (sisa kode untuk menghitung rata-rata IoU) ...
        # Setelah loop selesai, hitung rata-rata IoU
    average_iou = np.mean(iou_scores)
    print(f"\n==========================================")
    print(f"Rata-rata IoU (Mean IoU) dari seluruh dataset: {average_iou:.4f}")
    print(f"==========================================")
```

#Feature Extraction

```python
#GLCM saja

import numpy as np
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern, hog

# Asumsikan 'final_img' adalah list berisi gambar-gambar
glcm_features = []

# Loop melalui setiap gambar dalam list
for i in range(len(final_img)):
    # Ambil satu gambar dari list
    current_image = final_img[i]
    # Konversi tipe data gambar tunggal ini
    masked_img = current_image.astype(np.uint8)

    # Pastikan semua baris berikut ini memiliki indentasi yang sama

    # 1. GLCM Features
    # Hapus [i] dari sini -> gunakan 'masked_img' langsung
    glcm = graycomatrix(masked_img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
    glcm_props = [graycoprops(glcm, prop)[0, 0] for prop in ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM']]
    glcm_features.append(glcm_props)

# # Ubah list fitur menjadi matriks NumPy
# X = np.array(all_features)

# print("Ekstraksi fitur berhasil!")
# print(f"Bentuk matriks fitur (X): {X.shape}")
```

```python
#GLCM, + LBP

import numpy as np
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern, hog

# Asumsikan 'final_img' adalah list berisi gambar-gambar
glcm_lbp_features = []

# Loop melalui setiap gambar dalam list
for i in range(len(final_img)):
    # Ambil satu gambar dari list
    current_image = final_img[i]
    # Konversi tipe data gambar tunggal ini
    masked_img = current_image.astype(np.uint8)

    # Pastikan semua baris berikut ini memiliki indentasi yang sama

    # 1. GLCM Features
    # Hapus [i] dari sini -> gunakan 'masked_img' langsung
    glcm = graycomatrix(masked_img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
    glcm_props = [graycoprops(glcm, prop)[0, 0] for prop in ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM']]

    # 2. LBP Features
    # Hapus [i] dari sini
    lbp = local_binary_pattern(masked_img, P=8, R=1, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 11), range=(0, 10))
    lbp_hist = lbp_hist.astype("float")
    lbp_hist /= (lbp_hist.sum() + 1e-6)

    # Gabungkan semua fitur menjadi satu vektor
    combined_features = np.hstack([glcm_props, lbp_hist])
    glcm_lbp_features.append(combined_features)

# # Ubah list fitur menjadi matriks NumPy
# X = np.array(all_features)

# print("Ekstraksi fitur berhasil!")
# print(f"Bentuk matriks fitur (X): {X.shape}")
```

```python
#GLCM + HOG

import numpy as np
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern, hog

# Asumsikan 'final_img' adalah list berisi gambar-gambar
glcm_hog_features = []

# Loop melalui setiap gambar dalam list
for i in range(len(final_img)):
    # Ambil satu gambar dari list
    current_image = final_img[i]
    # Konversi tipe data gambar tunggal ini
    masked_img = current_image.astype(np.uint8)

    # Pastikan semua baris berikut ini memiliki indentasi yang sama

    # 1. GLCM Features
    # Hapus [i] dari sini -> gunakan 'masked_img' langsung
    glcm = graycomatrix(masked_img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
    glcm_props = [graycoprops(glcm, prop)[0, 0] for prop in ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM']]

    # 3. HOG Features
    # Hapus [i] dari sini
    hog_features = hog(masked_img, orientations=9, pixels_per_cell=(16, 16), cells_per_block=(2, 2), visualize=False)

    # Gabungkan semua fitur menjadi satu vektor
    combined_features = np.hstack([glcm_props, hog_features])
    glcm_hog_features.append(combined_features)

# # Ubah list fitur menjadi matriks NumPy
# X = np.array(all_features)

# print("Ekstraksi fitur berhasil!")
# print(f"Bentuk matriks fitur (X): {X.shape}")
```

```python
#GLCM, LBP, HOG

import numpy as np
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern, hog

# Asumsikan 'final_img' adalah list berisi gambar-gambar
all_features = []

# Loop melalui setiap gambar dalam list
for i in range(len(final_img)):
    # Ambil satu gambar dari list
    current_image = final_img[i]
    # Konversi tipe data gambar tunggal ini
    masked_img = current_image.astype(np.uint8)

    # Pastikan semua baris berikut ini memiliki indentasi yang sama

    # 1. GLCM Features
    # Hapus [i] dari sini -> gunakan 'masked_img' langsung
    glcm = graycomatrix(masked_img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
    glcm_props = [graycoprops(glcm, prop)[0, 0] for prop in ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM']]

    # 2. LBP Features
    # Hapus [i] dari sini
    lbp = local_binary_pattern(masked_img, P=8, R=1, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 11), range=(0, 10))
    lbp_hist = lbp_hist.astype("float")
    lbp_hist /= (lbp_hist.sum() + 1e-6)

    # 3. HOG Features
    # Hapus [i] dari sini
    hog_features = hog(masked_img, orientations=9, pixels_per_cell=(16, 16), cells_per_block=(2, 2), visualize=False)

    # Gabungkan semua fitur menjadi satu vektor
    combined_features = np.hstack([glcm_props, lbp_hist, hog_features])
    all_features.append(combined_features)

# # Ubah list fitur menjadi matriks NumPy
# X = np.array(all_features)

# print("Ekstraksi fitur berhasil!")
# print(f"Bentuk matriks fitur (X): {X.shape}")
```

```python
# GLCM + LBP + HOG + Statistik + Bentuk

# Impor library tambahan yang mungkin diperlukan
from skimage.measure import regionprops
from scipy.stats import skew, kurtosis
import pandas as pd # Menggunakan pandas untuk kemudahan

# --- ASUMSI ---
# 'list_isolated_lungs': list gambar grayscale paru-paru terisolasi (uint8)
# 'list_final_masks': list masker biner paru-paru (untuk fitur bentuk)

new_all_features = []

# Loop melalui setiap gambar
for i in range(len(final_img)):
    isolated_img = final_img[i]
    final_mask = masks_list[i]

    features = {}

    # 1. Fitur Statistik Orde Pertama
    pixels = isolated_img[isolated_img > 0]
    features['mean'] = np.mean(pixels)
    features['std'] = np.std(pixels)
    features['skew'] = skew(pixels)
    features['kurtosis'] = kurtosis(pixels)

    # 2. Fitur GLCM (Tekstur)
    glcm = graycomatrix(isolated_img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
    features['contrast'] = graycoprops(glcm, 'contrast')[0, 0]
    features['dissimilarity'] = graycoprops(glcm, 'dissimilarity')[0, 0]
    features['homogeneity'] = graycoprops(glcm, 'homogeneity')[0, 0]
    features['energy'] = graycoprops(glcm, 'energy')[0, 0]
    features['correlation'] = graycoprops(glcm, 'correlation')[0, 0]
    features['ASM'] = graycoprops(glcm, 'ASM')[0, 0]

    # 3. Fitur Bentuk (Shape)
    props = regionprops(final_mask)
    if props:
        areas = [p.area for p in props]
        perimeters = [p.perimeter for p in props]
        solidities = [p.solidity for p in props]
        eccentricities = [p.eccentricity for p in props]
        features['area_total'] = np.sum(areas)
        features['perimeter_mean'] = np.mean(perimeters)
        features['solidity_mean'] = np.mean(solidities)
        features['eccentricity_mean'] = np.mean(eccentricities)
    else:
        features.update({'area_total': 0, 'perimeter_mean': 0, 'solidity_mean': 0, 'eccentricity_mean': 0})

    # ---- TAMBAHKAN KEMBALI LBP DAN HOG ----

    # 4. Fitur LBP (Tekstur Mikro)
    lbp = local_binary_pattern(isolated_img, P=8, R=1, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 11), range=(0, 10))
    lbp_hist = lbp_hist.astype("float")
    lbp_hist /= (lbp_hist.sum() + 1e-6)
    for j, val in enumerate(lbp_hist):
        features[f'lbp_{j}'] = val # Menambahkan setiap bin histogram sebagai fitur terpisah

    # 5. Fitur HOG (Bentuk & Kontur)
    hog_features = hog(isolated_img, orientations=9, pixels_per_cell=(16, 16), cells_per_block=(2, 2), visualize=False)
    for j, val in enumerate(hog_features):
        features[f'hog_{j}'] = val # Menambahkan setiap elemen HOG sebagai fitur terpisah

    new_all_features.append(features)

# # Ubah list of dictionaries menjadi DataFrame, lalu menjadi NumPy array
# X_super_features = pd.DataFrame(all_features_list).fillna(0).values
# print(f"Ekstraksi 'super-feature' selesai. Bentuk matriks fitur (X) baru: {X_super_features.shape}")

# --- Lanjutkan ke tahap Split Data, Scaling, dan Training seperti biasa ---
```
