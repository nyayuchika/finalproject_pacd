```python
# Eksperimen Chan-Vese
# (Jalankan sekali saja di atas: install scikit-image)
!pip install scikit-image

import numpy as np
import cv2
from skimage.segmentation import chan_vese
from scipy import ndimage as ndi

# ---------------------------
# SEGMENTASI PARU DENGAN CHANâ€“VESE
# Input  : img_clahe (list/array citra grayscale hasil CLAHE)
# Output : cv_masks (list/array mask paru 0/255)
# ---------------------------

cv_masks = []

print("Running Chanâ€“Vese segmentation...")

for idx, img in enumerate(img_clahe):
    # Pastikan tipe data benar
    img_float = img.astype(np.float32) / 255.0

    # Chanâ€“Vese segmentation
    cv = chan_vese(
        img_float,
        mu=0.25,
        lambda1=1,
        lambda2=1,
        tol=1e-3,
        max_num_iter=200,
        dt=0.5,
        init_level_set="checkerboard",
    )

    # Ubah ke mask 0 / 255
    mask = (cv.astype(np.uint8) * 255)

    # Sedikit cleaning: fill holes dan smooth
    mask_filled = ndi.binary_fill_holes(mask).astype(np.uint8) * 255
    mask_smooth = cv2.medianBlur(mask_filled, 5)

    cv_masks.append(mask_smooth)

cv_masks = np.array(cv_masks)

print("Done. Total masks:", len(cv_masks))

# Cek satu contoh
i = 3  # boleh diganti indeks lain
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(plt.imread(list_images[i])); plt.title('Original')
plt.subplot(1,3,2); plt.imshow(img_clahe[i], cmap='gray'); plt.title('CLAHE')
plt.subplot(1,3,3); plt.imshow(cv_masks[i], cmap='gray'); plt.title('Chanâ€“Vese Mask')
plt.show()

```

```python
def otsu(image):
  #Membuat Histogram
  histogram = []
  for i in range(0, 256):
    nilai_awal = 0
    histogram.append(nilai_awal)

  #Hitung nilai intensitas tiap pixel
  for i in range(len(image)):
    for j in range(len(image[i])):
      nilai_pixel = image[i][j]
      histogram[nilai_pixel] = histogram[nilai_pixel] + 1

  #Hitung total pixel
  total = 0
  for i in range(256):
    total = total + histogram[i]

  best_threshold = 0
  min_within_class_var = float('inf')

  #Otsu dimulai
  for t in range (0, len(histogram)):
    #Background
    jumlah_histogram_b = 0
    for i in range(t):
      jumlah_histogram_b = jumlah_histogram_b + histogram[i]
    weight_b = jumlah_histogram_b / total
    jumlah_intensitas_b = 0
    jumlah_weight_b = 0
    for i in range(t):
      jumlah_intensitas_b = jumlah_intensitas_b + i * histogram[i]
      jumlah_weight_b = jumlah_weight_b + histogram[i]

    #Rata-rata
    if jumlah_weight_b == 0:
      mean_b = 0
    else:
      mean_b = jumlah_intensitas_b / jumlah_weight_b

    #Varians
    var_b = 0
    for i in range(t):
      var_b = var_b + ((i - mean_b) * (i - mean_b)) * histogram[i]
    if jumlah_weight_b != 0:
      var_b = var_b / jumlah_weight_b

    #------------------------------------------------------#

    #Foreground
    jumlah_histogram_f = 0
    for i in range(t, len(histogram)):
      jumlah_histogram_f = jumlah_histogram_f + histogram[i]
    weight_f = jumlah_histogram_f / total
    jumlah_intensitas_f = 0
    jumlah_weight_f = 0
    for i in range(t, len(histogram)):
      jumlah_intensitas_f = jumlah_intensitas_f + i * histogram[i]
      jumlah_weight_f = jumlah_weight_f + histogram[i]

    #Rata-rata
    if jumlah_weight_f == 0:
      mean_f = 0
    else:
      mean_f = jumlah_intensitas_f / jumlah_weight_f

    #Varians
    var_f = 0
    for i in range(t, len(histogram)):
      var_f = var_f + ((i - mean_f) * (i - mean_f)) * histogram[i]
    if jumlah_weight_f != 0:
      var_f = var_f / jumlah_weight_f

    #Within class variance
    within_class_var = (weight_b * var_b) + (weight_f * var_f)

    #Pilih nilai threshold minimum
    if within_class_var < min_within_class_var:
      min_within_class_var = within_class_var
      best_threshold = t
    print(f"t={t} | Wb={weight_b:.6f} | Wf={weight_f:.6f} | Î¼b={mean_b:.6f} | Î¼f={mean_f:.6f} | ÏƒÂ²w={within_class_var:.6f}")

  print("\nNilai threshold terbaik =", best_threshold)

  #Menerapkan otsu ke citra
  # otsu_image = (image >= best_threshold) * 255
  otsu_image = (image < best_threshold) * 255 #untuk hasil invers
  return otsu_image, best_threshold
```

```python
from scipy import ndimage
import cv2
import numpy as np

def morpho_close_pipeline(img_list, kernel_size):
    """Melakukan morphological closing + binary fill + ambil 2 komponen paru terbesar"""
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    # kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (kernel_size,kernel_size))

    result_list = []

    for img in img_list:
        # Pastikan tipe data uint8
        img = img.astype(np.uint8)

        # Tahap 1: morphological closing
        closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

        # Tahap 2: isi lubang internal
        filled = ndimage.binary_fill_holes(closing).astype(np.uint8) * 255

        # Tahap 3: ambil 2 area paru terbesar
        num_labels, labels = cv2.connectedComponents(filled)
        sizes = [np.sum(labels == i) for i in range(1, num_labels)]
        largest = sorted(range(1, num_labels), key=lambda i: sizes[i-1], reverse=True)[:2]
        mask_clean = np.isin(labels, largest).astype(np.uint8) * 255

        # Tahap 4: haluskan tepi
        mask_clean = cv2.medianBlur(mask_clean, 5)

        result_list.append(mask_clean)

    return np.asarray(result_list)


# === Eksekusi dua tahap closing ===
# Tahap pertama: kernel 9x9
img_morpho_close1 = morpho_close_pipeline(flood_fill_list, kernel_size=9)

# Tahap kedua: kernel 11x11
img_morpho_close2 = morpho_close_pipeline(img_morpho_close1, kernel_size=15)

# # Tahap ketiga: kernel 15x15
# img_morpho_close2 = morpho_close_pipeline(img_morpho_close1, kernel_size=15)

```

```python
import numpy as np

def calculate_iou(mask_prediksi, mask_ground_truth):
    """
    Menghitung Intersection over Union (IoU) antara dua masker biner.

    Args:
    - mask_prediksi: NumPy array dari masker hasil segmentasi Anda (nilai 0 atau 255).
    - mask_ground_truth: NumPy array dari masker kunci jawaban (nilai 0 atau 255).

    Returns:
    - iou_score: Nilai IoU (float antara 0 dan 1).
    """
    # Pastikan kedua masker memiliki tipe data yang sama dan biner (0 atau 1)
    # untuk operasi logika. Kita ubah ke boolean (True/False).
    mask_pred_bool = mask_prediksi.astype(bool)
    mask_ground_truth_bool = mask_ground_truth.astype(bool)

    # Hitung Intersection (area di mana KEDUA masker True)
    intersection = np.logical_and(mask_pred_bool, mask_ground_truth_bool)

    # Hitung Union (area di mana SALAH SATU masker True)
    union = np.logical_or(mask_pred_bool, mask_ground_truth_bool)

    # Hitung jumlah piksel di intersection dan union
    intersection_area = np.sum(intersection)
    union_area = np.sum(union)

    # Hitung IoU, tambahkan epsilon (nilai kecil) untuk menghindari pembagian dengan nol
    iou_score = intersection_area / (union_area + 1e-6)

    return iou_score
```

```python
#GLCM, LBP, HOG

import numpy as np
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern, hog

# Asumsikan 'final_img' adalah list berisi gambar-gambar
all_features = []

# Loop melalui setiap gambar dalam list
for i in range(len(final_img)):
    # Ambil satu gambar dari list
    current_image = final_img[i]
    # Konversi tipe data gambar tunggal ini
    masked_img = current_image.astype(np.uint8)

    # Pastikan semua baris berikut ini memiliki indentasi yang sama

    # 1. GLCM Features
    # Hapus [i] dari sini -> gunakan 'masked_img' langsung
    glcm = graycomatrix(masked_img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
    glcm_props = [graycoprops(glcm, prop)[0, 0] for prop in ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM']]

    # 2. LBP Features
    # Hapus [i] dari sini
    lbp = local_binary_pattern(masked_img, P=8, R=1, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 11), range=(0, 10))
    lbp_hist = lbp_hist.astype("float")
    lbp_hist /= (lbp_hist.sum() + 1e-6)

    # 3. HOG Features
    # Hapus [i] dari sini
    hog_features = hog(masked_img, orientations=9, pixels_per_cell=(16, 16), cells_per_block=(2, 2), visualize=False)

    # Gabungkan semua fitur menjadi satu vektor
    combined_features = np.hstack([glcm_props, lbp_hist, hog_features])
    all_features.append(combined_features)

# # Ubah list fitur menjadi matriks NumPy
# X = np.array(all_features)

# print("Ekstraksi fitur berhasil!")
# print(f"Bentuk matriks fitur (X): {X.shape}")
```

```python
# GLCM + LBP + HOG + Statistik + Bentuk

# Impor library tambahan yang mungkin diperlukan
from skimage.measure import regionprops
from scipy.stats import skew, kurtosis
import pandas as pd # Menggunakan pandas untuk kemudahan

# --- ASUMSI ---
# 'list_isolated_lungs': list gambar grayscale paru-paru terisolasi (uint8)
# 'list_final_masks': list masker biner paru-paru (untuk fitur bentuk)

new_all_features = []

# Loop melalui setiap gambar
for i in range(len(final_img)):
    isolated_img = final_img[i]
    final_mask = masks_list[i]

    features = {}

    # 1. Fitur Statistik Orde Pertama
    pixels = isolated_img[isolated_img > 0]
    features['mean'] = np.mean(pixels)
    features['std'] = np.std(pixels)
    features['skew'] = skew(pixels)
    features['kurtosis'] = kurtosis(pixels)

    # 2. Fitur GLCM (Tekstur)
    glcm = graycomatrix(isolated_img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
    features['contrast'] = graycoprops(glcm, 'contrast')[0, 0]
    features['dissimilarity'] = graycoprops(glcm, 'dissimilarity')[0, 0]
    features['homogeneity'] = graycoprops(glcm, 'homogeneity')[0, 0]
    features['energy'] = graycoprops(glcm, 'energy')[0, 0]
    features['correlation'] = graycoprops(glcm, 'correlation')[0, 0]
    features['ASM'] = graycoprops(glcm, 'ASM')[0, 0]

    # 3. Fitur Bentuk (Shape)
    props = regionprops(final_mask)
    if props:
        areas = [p.area for p in props]
        perimeters = [p.perimeter for p in props]
        solidities = [p.solidity for p in props]
        eccentricities = [p.eccentricity for p in props]
        features['area_total'] = np.sum(areas)
        features['perimeter_mean'] = np.mean(perimeters)
        features['solidity_mean'] = np.mean(solidities)
        features['eccentricity_mean'] = np.mean(eccentricities)
    else:
        features.update({'area_total': 0, 'perimeter_mean': 0, 'solidity_mean': 0, 'eccentricity_mean': 0})

    # ---- TAMBAHKAN KEMBALI LBP DAN HOG ----

    # 4. Fitur LBP (Tekstur Mikro)
    lbp = local_binary_pattern(isolated_img, P=8, R=1, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 11), range=(0, 10))
    lbp_hist = lbp_hist.astype("float")
    lbp_hist /= (lbp_hist.sum() + 1e-6)
    for j, val in enumerate(lbp_hist):
        features[f'lbp_{j}'] = val # Menambahkan setiap bin histogram sebagai fitur terpisah

    # 5. Fitur HOG (Bentuk & Kontur)
    hog_features = hog(isolated_img, orientations=9, pixels_per_cell=(16, 16), cells_per_block=(2, 2), visualize=False)
    for j, val in enumerate(hog_features):
        features[f'hog_{j}'] = val # Menambahkan setiap elemen HOG sebagai fitur terpisah

    new_all_features.append(features)

# # Ubah list of dictionaries menjadi DataFrame, lalu menjadi NumPy array
# X_super_features = pd.DataFrame(all_features_list).fillna(0).values
# print(f"Ekstraksi 'super-feature' selesai. Bentuk matriks fitur (X) baru: {X_super_features.shape}")

# --- Lanjutkan ke tahap Split Data, Scaling, dan Training seperti biasa ---
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# --- ASUMSI ---
# 1. 'all_features_list': List berisi DICTIONARY fitur Anda.
# 2. 'y': List atau array berisi LABEL yang urutannya sesuai.

# =============================================================================
# Tahap 1: Konversi List of Dictionaries menjadi Matriks Fitur (X)
# =============================================================================
# Gunakan Pandas untuk mengubah list of dicts menjadi DataFrame,
# lalu ambil nilainya sebagai matriks NumPy.
# .fillna(0) untuk mengisi nilai yang hilang jika ada (misal, fitur bentuk gagal dihitung)
x = pd.DataFrame(new_all_features).fillna(0).values

print(f"Bentuk matriks fitur mentah (X): {x.shape}")


# =============================================================================
# Tahap 2: Split Data (Split Dulu!)
# =============================================================================
# Gunakan matriks X yang sudah jadi untuk di-split
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print(f"Bentuk X_train: {x_train.shape}")
print(f"Bentuk X_test: {x_test.shape}")


# =============================================================================
# Tahap 3: Scaling Fitur (Baru Scaling)
# =============================================================================
# Buat objek scaler
scaler = MinMaxScaler()

# 'fit' scaler HANYA pada data latih untuk mempelajari skala
scaler.fit(x_train)

# Terapkan ('transform') scaler pada kedua set data
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)

print("\nScaling selesai.")
print(f"Nilai minimal X_train_scaled: {x_train_scaled.min():.2f}")
print(f"Nilai maksimal X_train_scaled: {x_train_scaled.max():.2f}")


# --- SELESAI ---
# Variabel X_train_scaled dan X_test_scaled sekarang siap digunakan
# untuk melatih dan menguji model machine learning Anda.
```

###SVM

```python
from sklearn.svm import SVC

model = SVC(kernel='rbf',class_weight='balanced')
model.fit(x_train_scaled, y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred, target_names=['Normal','TBC']))
```

###KNN

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
# Latih model menggunakan data latih
model.fit(x_train_scaled, y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```
