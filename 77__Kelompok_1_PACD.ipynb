```python
#Otsu
def otsu(image):
  #Membuat Histogram
  histogram = []
  for i in range(0, 256):
    nilai_awal = 0
    histogram.append(nilai_awal)

  #Hitung nilai intensitas tiap pixel
  for i in range(len(image)):
    for j in range(len(image[i])):
      nilai_pixel = image[i][j]
      histogram[nilai_pixel] = histogram[nilai_pixel] + 1

  #Hitung total pixel
  total = 0
  for i in range(256):
    total = total + histogram[i]

  best_threshold = 0
  min_within_class_var = float('inf')

  #Otsu dimulai
  for t in range (0, len(histogram)):
    #Background
    jumlah_histogram_b = 0
    for i in range(t):
      jumlah_histogram_b = jumlah_histogram_b + histogram[i]
    weight_b = jumlah_histogram_b / total
    jumlah_intensitas_b = 0
    jumlah_weight_b = 0
    for i in range(t):
      jumlah_intensitas_b = jumlah_intensitas_b + i * histogram[i]
      jumlah_weight_b = jumlah_weight_b + histogram[i]

    #Rata-rata
    if jumlah_weight_b == 0:
      mean_b = 0
    else:
      mean_b = jumlah_intensitas_b / jumlah_weight_b

    #Varians
    var_b = 0
    for i in range(t):
      var_b = var_b + ((i - mean_b) * (i - mean_b)) * histogram[i]
    if jumlah_weight_b != 0:
      var_b = var_b / jumlah_weight_b

    #------------------------------------------------------#

    #Foreground
    jumlah_histogram_f = 0
    for i in range(t, len(histogram)):
      jumlah_histogram_f = jumlah_histogram_f + histogram[i]
    weight_f = jumlah_histogram_f / total
    jumlah_intensitas_f = 0
    jumlah_weight_f = 0
    for i in range(t, len(histogram)):
      jumlah_intensitas_f = jumlah_intensitas_f + i * histogram[i]
      jumlah_weight_f = jumlah_weight_f + histogram[i]

    #Rata-rata
    if jumlah_weight_f == 0:
      mean_f = 0
    else:
      mean_f = jumlah_intensitas_f / jumlah_weight_f

    #Varians
    var_f = 0
    for i in range(t, len(histogram)):
      var_f = var_f + ((i - mean_f) * (i - mean_f)) * histogram[i]
    if jumlah_weight_f != 0:
      var_f = var_f / jumlah_weight_f

    #Within class variance
    within_class_var = (weight_b * var_b) + (weight_f * var_f)

    #Pilih nilai threshold minimum
    if within_class_var < min_within_class_var:
      min_within_class_var = within_class_var
      best_threshold = t
    print(f"t={t} | Wb={weight_b:.6f} | Wf={weight_f:.6f} | b={mean_b:.6f} | f={mean_f:.6f} | w={within_class_var:.6f}")

  print("\nNilai threshold terbaik =", best_threshold)

  #Menerapkan otsu ke citra
  # otsu_image = (image >= best_threshold) * 255
  otsu_image = (image < best_threshold) * 255 #untuk hasil invers
  return otsu_image, best_threshold
```

```python
from scipy import ndimage
import cv2
import numpy as np

def morpho_close_pipeline(img_list, kernel_size):
  kernel = np.ones((kernel_size, kernel_size), np.uint8)
  result_list = []

  for img in img_list:
    img = img.astype(np.uint8)
    #Morphological closing
    closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
    #isi lubang
    filled = ndimage.binary_fill_holes(closing).astype(np.uint8) * 255
    #ambil 2 area paru terbesar
    num_labels, labels = cv2.connectedComponents(filled)
    sizes = [np.sum(labels == i) for i in range(1, num_labels)]
    largest = sorted(range(1, num_labels), key=lambda i: sizes[i-1], reverse=True)[:2]
    mask_clean = np.isin(labels, largest).astype(np.uint8) * 255
    #haluskan tepi
    mask_clean = cv2.medianBlur(mask_clean, 5)
    result_list.append(mask_clean)
  return np.asarray(result_list)


# closing 2 tahap
# Tahap pertama: kernel 9x9
img_morpho_close1 = morpho_close_pipeline(flood_fill_list, kernel_size=9)

# Tahap kedua: kernel 11x11
img_morpho_close2 = morpho_close_pipeline(img_morpho_close1, kernel_size=15)
```

```python
final_img = []
for i in range(len(img_morpho_close2)):
  image = img_morpho_close2[i]
  lung_mask_filled = ndimage.binary_fill_holes(image).astype(np.uint8) * 255
  num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(lung_mask_filled, connectivity=8)

  final_mask = np.zeros_like(image)
  if num_labels > 1:
    areas = stats[1:, cv2.CC_STAT_AREA]
    sorted_indices = np.argsort(areas)[::-1]
    num_components_to_keep = min(2, len(sorted_indices))
    largest_indices = sorted_indices[:num_components_to_keep] + 1
    final_mask = np.isin(labels, largest_indices).astype(np.uint8) * 255
    final_mask = cv2.medianBlur(final_mask, 5)
  clahe_image = img_clahe[i]
  isolated_lungs = cv2.bitwise_and(clahe_image, clahe_image, mask=final_mask)
  final_img.append(isolated_lungs)
```

```python
#Hitung IoU
import numpy as np

def calculate_iou(mask_prediksi, mask_ground_truth):
  mask_pred_bool = mask_prediksi.astype(bool)
  mask_ground_truth_bool = mask_ground_truth.astype(bool)
  intersection = np.logical_and(mask_pred_bool, mask_ground_truth_bool)
  union = np.logical_or(mask_pred_bool, mask_ground_truth_bool)
  intersection_area = np.sum(intersection)
  union_area = np.sum(union)
  iou_score = intersection_area / (union_area + 1e-6)
  return iou_score
```

```python
#Feature Extraction
#GLCM, LBP, HOG, Sobel
import numpy as np
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern, hog

all_features = []
for i in range(len(final_img)):
  current_image = final_img[i]
  masked_img = current_image.astype(np.uint8)
  #1. GLCM Features
  angles = [
      np.pi/4,      # 45
      np.pi/2,      # 90
      3*np.pi/4,    # 135
      5*np.pi/4,    # 225
      3*np.pi/2,    # 270
      7*np.pi/4     # 315
  ]
  glcm = graycomatrix(masked_img, distances=[1], angles=angles, levels=256, symmetric=True, normed=True)
  glcm_props = [graycoprops(glcm, prop)[0, 0] for prop in ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM']]

  #2. LBP Features
  lbp = local_binary_pattern(masked_img, P=12, R=2, method='uniform')
  lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 11), range=(0, 10))
  lbp_hist = lbp_hist.astype("float")
  lbp_hist /= (lbp_hist.sum() + 1e-6)

  #3. HOG Features
  hog_features = hog(masked_img, orientations=9, pixels_per_cell=(16, 16), cells_per_block=(2, 2), visualize=False)

  #4. SOBEL (fitur serat bronkovaskular)
  sobel_x = cv2.Sobel(masked_img, cv2.CV_64F, dx=1, dy=0, ksize=3)
  sobel_y = cv2.Sobel(masked_img, cv2.CV_64F, dx=0, dy=1, ksize=3)
  #Magnitude
  sobel_mag = np.sqrt(sobel_x**2 + sobel_y**2)
  sobel_mag_mean = np.mean(sobel_mag)
  sobel_mag_std  = np.std(sobel_mag)

  combined_features = np.hstack([glcm_props, lbp_hist, hog_features, [sobel_mag_mean, sobel_mag_std]])
  all_features.append(combined_features)
```

```python
import numpy as np

y_labels = []
print("Sedang membuat label dari list_masks...")
for path in list_masks:
  path_lower = str(path).lower()
  if "/normal/" in path_lower:
    y_labels.append(0)  #0 untuk Normal
  elif "/tbc/" in path_lower:
    y_labels.append(1)  #1 untuk TBC
  else:
    print(f"Warning: Tidak bisa menentukan label untuk file: {path}")

y = np.array(y_labels)

print("Selesai!")
print(f"Total Label: {len(y)}")
print(f"Jumlah Normal (0): {np.sum(y == 0)}")
print(f"Jumlah TBC (1): {np.sum(y == 1)}")
```

```python
x = np.array(all_features)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
print(f"Bentuk X_train: {x_train.shape}")
print(f"Bentuk X_test: {x_test.shape}")

#Normalisasi
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
print("\nScaling selesai.")
print(f"Nilai minimal X_train_scaled: {x_train_scaled.min():.2f}")
print(f"Nilai maksimal X_train_scaled: {x_train_scaled.max():.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.svm import SVC  # Ganti RF ke SVM
from sklearn.model_selection import GridSearchCV # Alat untuk tuning
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#PCA
print("\n--- 1. Konfigurasi & Fitting PCA ---")
pca = PCA(n_components=0.5)
x_train_pca = pca.fit_transform(x_train_scaled)
x_test_pca = pca.transform(x_test_scaled)
num_components = x_train_pca.shape[1]
print(f"Jumlah fitur (dimensi) awal: {x_train_scaled.shape[1]}")
print(f"Jumlah komponen PCA yang terbentuk: {num_components}")
print(f"Total variance (informasi) yang dijaga: {np.sum(pca.explained_variance_ratio_):.2%}")
print(f"Variance ratio 5 komponen pertama: {pca.explained_variance_ratio_[:5]}")

#SVM
print("\n--- 3. Training Model (SVM dengan Tuning) ---")
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'linear', 'poly']
}

print("Sedang melakukan tuning parameter SVM... Mohon tunggu.")
grid = GridSearchCV(
    SVC(class_weight='balanced'),
    param_grid,
    refit=True,  #fit ulang model dengan param terbaik di akhir
    verbose=1,
    cv=3
)

grid.fit(x_train_pca, y_train)
print(f"\nParameter Terbaik SVM: {grid.best_params_}")
print(f"Score Terbaik (saat training): {grid.best_score_:.4f}")
y_pred = grid.best_estimator_.predict(x_test_pca)

print("\n--- Hasil Klasifikasi dengan PCA + Tuned SVM ---")
print(classification_report(y_test, y_pred))

plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Jumlah Komponen')
plt.ylabel('Cumulative Explained Variance')
plt.title('Seberapa banyak informasi yang dijelaskan oleh Komponen PCA')
plt.grid(True)
plt.savefig('pca_variance_plot.png')
print("Grafik variance telah disimpan sebagai 'pca_variance_plot.png'")
```

```python
import cv2
import numpy as np
from google.colab import files
from PIL import Image
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern, hog
from scipy import ndimage

uploaded = files.upload()

file_name = list(uploaded.keys())[0]
img = cv2.imread(file_name)

# Resize
img = cv2.resize(img, (256,256), interpolation=cv2.INTER_LINEAR)

# Convert ke grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Gaussian blur
gauss = cv2.GaussianBlur(gray, (3,3), 0)

# CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
clahe_img = clahe.apply(gauss)

# Otsu versi OpenCV (lebih cepat)
_, otsu_bin = cv2.threshold(clahe_img, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Flood fill pinggiran
padded = cv2.copyMakeBorder(otsu_bin, 1,1,1,1, cv2.BORDER_CONSTANT, value=255)
mask = np.zeros((padded.shape[0]+2, padded.shape[1]+2), np.uint8)
cv2.floodFill(padded, mask, (0,0), 0)
flooded = padded[1:-1,1:-1]

# Morphological closing
kernel = np.ones((11,11), np.uint8)
closed = cv2.morphologyEx(flooded, cv2.MORPH_CLOSE, kernel)

# Fill holes
filled = ndimage.binary_fill_holes(closed).astype(np.uint8)*255

# Ambil 2 komponen paru terbesar
num_labels, labels = cv2.connectedComponents(filled)
areas = [(labels == i).sum() for i in range(1, num_labels)]
largest = np.argsort(areas)[-2:] + 1
lung_mask = np.isin(labels, largest).astype(np.uint8)*255

# Final masked image
isolated_lung = cv2.bitwise_and(clahe_img, clahe_img, mask=lung_mask)

masked_img = isolated_lung.astype(np.uint8)

# GLCM
angles = [np.pi/4, np.pi/2, 3*np.pi/4, 5*np.pi/4, 3*np.pi/2, 7*np.pi/4]
glcm = graycomatrix(masked_img, distances=[1], angles=angles, levels=256, symmetric=True, normed=True)
glcm_props = [graycoprops(glcm, prop)[0,0] for prop in
              ['contrast','dissimilarity','homogeneity','energy','correlation','ASM']]

# LBP
lbp = local_binary_pattern(masked_img, P=12, R=2, method='uniform')
lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0,11), range=(0,10))
lbp_hist = lbp_hist.astype("float")
lbp_hist /= (lbp_hist.sum()+1e-6)

# HOG
hog_features = hog(masked_img, orientations=9, pixels_per_cell=(16,16),
                   cells_per_block=(2,2), visualize=False)

# Sobel
sobel_x = cv2.Sobel(masked_img, cv2.CV_64F, 1,0,ksize=3)
sobel_y = cv2.Sobel(masked_img, cv2.CV_64F, 0,1,ksize=3)
sobel_mag = np.sqrt(sobel_x**2 + sobel_y**2)

sobel_mean = np.mean(sobel_mag)
sobel_std  = np.std(sobel_mag)

# Gabungkan semua fitur
feature_vector = np.hstack([glcm_props, lbp_hist, hog_features, [sobel_mean, sobel_std]])
feature_vector = feature_vector.reshape(1, -1)

import joblib
scaler = joblib.load("model_scaler.pkl")
pca    = joblib.load("model_pca.pkl")
svm    = joblib.load("model_svm.pkl")

#Scaling
scaled = scaler.transform(feature_vector)

#PCA
reduced = pca.transform(scaled)

#Prediksi
pred = svm.predict(reduced)[0]

label = "NORMAL" if pred == 0 else "TBC"
print("\n===== HASIL PREDIKSI =====")
print("Hasil klasifikasi gambar:", label)
```
